```python
max_search=int(3e5)

def mask_arr(id):
    mask = np.full((max_search,1), False)
    mask[id-1] = True
    return mask

persistent_ids = np.full((max_search,1), True)

for i in range(len(ids)):
    persistent_ids = persistent_ids * mask_arr(ids[i])

print(sum(persistent_ids))
```
